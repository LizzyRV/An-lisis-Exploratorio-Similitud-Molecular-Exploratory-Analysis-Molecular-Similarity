Percepción de similitud molecular utilizando modelos de aprendizaje automático. El documento propone herramientas basadas en aprendizaje automático que pueden complementar el juicio humano en la evaluación de similitud molecular, lo que podría ser útil en la investigación y regulación de medicamentos.

La similitud molecular es un concepto clave en muchas áreas de la química, especialmente en la industria farmacéutica. El principio general es que "moléculas similares tienden a tener propiedades similares", lo que ha llevado a la creación de métodos para evaluar la similitud molecular, que son ampliamente utilizados en estudios de relaciones estructura-actividad y en la regulación de medicamentos.

El estudio se centra en la creación de modelos de aprendizaje automático que imitan la percepción humana de similitud molecular. Los investigadores utilizaron tanto descriptores 2D (huellas moleculares) como descriptores 3D (forma molecular y farmacóforos) para construir estos modelos. Se evaluaron en un conjunto de datos donde un grupo de expertos dio su opinión sobre la similitud entre pares de moléculas. El objetivo es que estos modelos asistan en la toma de decisiones en contextos como la investigación de nuevos medicamentos o la evaluación de la similitud de moléculas en el contexto de la legislación de medicamentos huérfanos.

Aplicación en fármacos huérfanos: Estos medicamentos son utilizados para tratar enfermedades raras y debido a su pequeño mercado, las agencias regulatorias ofrecen incentivos para su desarrollo.
Uno de los criterios clave es si un nuevo fármaco es lo suficientemente diferente de los ya existentes.

Modelos basados en huellas moleculares 2D y 3D: Los modelos creados utilizan la métrica de Tanimoto para comparar moléculas en 2D y en 3D, basándose en la estructura molecular y su alineación.

Encuesta de expertos: Se recopiló un conjunto de datos donde expertos evaluaron la similitud de pares de moléculas, utilizando tanto representaciones 2D como 3D.

Estos datos se usaron para entrenar los modelos de aprendizaje automático.
Aplicación de los modelos: Los modelos pueden predecir la probabilidad de que un panel de expertos considere que dos moléculas son similares, y se han implementado como una herramienta en línea de acceso público.

Los modelos no reemplazan al ojo humano, pero mejoran la eficiencia al proporcionar un filtro inicial que puede enfocarse en casos donde la decisión es más difícil o donde la similitud es debatible entre los expertos.

El coeficiente de Tanimoto es una métrica que mide la similitud entre dos conjuntos. En el contexto de la química computacional y bioinformática, se utiliza principalmente para medir la similitud entre moléculas, especialmente comparando sus huellas moleculares o descriptores.

Una huella molecular es una representación binaria de una molécula, en la que cada bit representa una característica de la estructura química, como la presencia o ausencia de ciertos átomos, grupos funcionales, enlaces, o patrones de subestructuras. Se crea para cada molécula un vector que describe estas características.

Aplicaciones del coeficiente de Tanimoto:
Cribado virtual (Virtual Screening): Se utiliza para encontrar moléculas con estructuras similares a un compuesto objetivo, lo cual es útil en el diseño de fármacos.
Comparación molecular: Es una herramienta clave en la química computacional para agrupar o clasificar moléculas según su similitud estructural.
Predicción de propiedades: Se usa para inferir que moléculas con altos valores de Tanimoto pueden tener propiedades físicas, químicas o biológicas similares.

Tanimoto 2D: Se calcula utilizando las huellas moleculares basadas en la estructura 2D de las moléculas, que representan características como la conectividad de los átomos.
Tanimoto 3D (TanimotoCombo): En este caso, se consideran descriptores tridimensionales como la forma y la orientación de las moléculas, para capturar información sobre cómo las moléculas interactuarían en un espacio tridimensional.

Huella molecular:

Es una representación computacional de una molécula pequeña, como un ligando o un compuesto químico.
Se utiliza para comparar moléculas entre sí, describiendo la presencia de ciertos grupos funcionales, patrones de átomos, enlaces, o características químicas de la estructura.
En lugar de usar la estructura completa de la molécula, se convierte en una secuencia de bits (1s y 0s), donde cada bit indica la presencia o ausencia de una característica específica de la molécula.
Se aplica a compuestos pequeños, como los que interactúan con proteínas en el contexto del diseño de fármacos.

1. TanimotoCombo (Similitud combinada 3D)
El TanimotoCombo es una métrica que evalúa la similitud tridimensional (3D) entre dos moléculas. A diferencia de las huellas moleculares 2D, que se basan en la conectividad de los átomos en un plano, la métrica TanimotoCombo compara la forma 3D y los grupos farmacóforos de las moléculas.

¿Cómo funciona?
Forma 3D: Cada molécula tiene una estructura tridimensional que influye en cómo interactúa con otras moléculas (por ejemplo, una proteína). ROCS (Rapid Overlay of Chemical Structures) es una herramienta que compara estas formas tridimensionales para ver qué tan bien se alinean las moléculas en el espacio.
Grupos farmacóforos: Estos son sitios dentro de una molécula que son importantes para la actividad biológica, como un sitio que puede formar enlaces con una proteína. ROCS también tiene en cuenta cómo se distribuyen estos grupos en el espacio tridimensional.
¿Qué mide TanimotoCombo?
El TanimotoCombo mide la superposición entre las formas 3D de las dos moléculas, así como la orientación espacial de sus grupos funcionales. Básicamente, evalúa cuán bien las dos moléculas se alinean en tres dimensiones.

Un valor alto de TanimotoCombo (cercano a 2.0) indica que las moléculas tienen formas y distribuciones espaciales similares.
Un valor bajo indica que las moléculas son muy diferentes en términos de su forma y estructura 3D.
Este tipo de comparación es particularmente útil en estudios de diseño de fármacos, donde la forma tridimensional de una molécula afecta cómo interactúa con una proteína objetivo.

 Frac_similar (Fracción de expertos que consideran las moléculas similares)
La columna frac_similar en tu dataset refleja la opinión de un panel de expertos humanos sobre si consideran que dos moléculas son similares o no. Este valor es una fracción (entre 0 y 1) que indica el porcentaje de expertos que evaluaron las moléculas como similares.

Qué significa este valor=
Frac_similar es el resultado de encuestas o evaluaciones cualitativas realizadas por humanos, generalmente expertos en química o bioinformática.
Si frac_similar = 1, significa que el 100% de los expertos consideraron que las dos moléculas son similares.
Si frac_similar = 0, significa que ninguno de los expertos consideró que las moléculas son similares.
Valores intermedios (por ejemplo, 0.5) indican que una parte del panel de expertos las consideró similares, mientras que la otra no.
¿Por qué es útil?
Este dato es útil porque, a pesar de los cálculos computacionales de similitud (como el coeficiente de Tanimoto o TanimotoCombo), la percepción humana a veces puede diferir, especialmente en casos ambiguos o donde los algoritmos no capturan bien ciertas características moleculares. La fracción de expertos proporciona una medida cualitativa que complementa los cálculos de similitud automáticos.


TanimotoCombo mide qué tan bien se alinean las moléculas en el espacio tridimensional en términos de su forma y grupos funcionales.
Frac_similar es la fracción de expertos humanos que consideran que dos moléculas son similares, proporcionándote una visión más subjetiva basada en el criterio humano.
Ambas columnas son valiosas para estudiar cómo la similitud computacional (TanimotoCombo) se correlaciona con la percepción humana de similitud (frac_similar), lo que es útil para validar modelos predictivos o identificar casos en los que los modelos no coinciden con la evaluación humana.

```python
from rdkit import Chem
from rdkit.Chem import Descriptors
```

RDKit tiene la función Chem.MolFromSmiles() que convierte una cadena SMILES en un objeto de molécula.
Si falla (es decir, la cadena SMILES es inválida), devuelve None o lanza un error.
Esta función se aplica después a todas las cadenas SMILES

```python
def validate_smiles(smiles):
  try:
    Chem.MolFromSmiles(smiles)
    return True
  except:
    return False

# Verificar la validez de las moléculas SMILES en ambas columnas
data['valid_molecule_a'] = data['curated_smiles_molecule_a'].apply(validate_smiles)
data['valid_molecule_b'] = data['curated_smiles_molecule_b'].apply(validate_smiles)

# Resumen de moléculas inválidas
invalid_molecule_a = data[~data['valid_molecule_a']]
invalid_molecule_b = data[~data['valid_molecule_b']]

# Guardar resultados de moléculas inválidas
invalid_molecule_a.to_csv('invalid_molecule_a.csv', index=False)
invalid_molecule_b.to_csv('invalid_molecule_b.csv', index=False)
```
